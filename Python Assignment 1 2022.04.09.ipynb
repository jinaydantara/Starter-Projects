```python
## replace "years" with "" in the age
## similarly replacing "$" with ""


retention['age'] = retention['age'].str.replace('years','',regex = True)
retention['gross_income'] = retention['gross_income'].str.replace('$','',regex = True)

## renaming columns that had unclear column names
## tiprel_1mes is renamed to inactivity_begn
## ind_actividad_cliente is renamed to inactivity_end

retention.rename(columns={'tiprel_1mes':'inactivity_begn'},inplace = True)
retention.rename(columns={'ind_actividad_cliente':'inactivity_end'}, inplace = True)

retention.head()
```

```python
## to make the change permanent, pulled it to another csv
## need to check if there is a better / faster way to do this

retention.to_csv('df1.csv')
```

## Data Exploration

Now that the data is cleaned (to the best of my ability), moving to Task 2 of the project

```python
df2.groupby(by = ['gender','segment']).agg({'gross_income' : 'mean'}).round(2)
```

```python
## creating the pivot table to analyze the data

P1 = pd.pivot_table(df2, index = ['inactivity_begn','inactivity_end','gender'],
                    values = ['age','gross_income','num_credit_cards','num_Products'],
               aggfunc = ['mean','median','count']).round(2)
P1.style.format('{:,}')
```

Here I am identifying the buckets as follows:

1.   Inactive at begin | inactive at end
2.   Inactive at  begin | active at end
1.   Active at begin | inactive at end
2.   Active at begin | active at end

We see that the customers with the lowest annual income of the 4 buckets have changed their behaviour i.e. inactive to active and they seem to be using 1 product of the bank

```python
P1.query('inactivity_end == 1').style.format('{:,}')
```

People wih high income tend to be active. They seem to be using more products as well compared to the group that got inactive

Compared to the pivot table above as well it seems that the group with the highest annual income are more active in terms of using banking products. Females seem to be using more products on average compared to Males in this bucket

##Analysis of behaviour
